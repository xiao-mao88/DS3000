(7 pts) Extract data from 2003 to 2018 (inclusive), for all passengers who deplaned the aircraft in May e.g. May 2003, May 2004, May 2005 etc. Visualize the extracted data, using a line chart, and comment on the trend.

```python
#extract the years within the range & from May
extracted_data = airTrafficPassengerStats_df[
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.to_period('Y') <= '2018') &
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.to_period('Y') >= '2003') &
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.month == 5)]
#check that data has been extracted correctly
extracted_data.head()
```

```python
#get deplane, group by year, aggregate data
final_extraction = extracted_data[(extracted_data['Activity Type Code'] == 'Deplaned')].groupby(extracted_data['Activity Period Start Date'].dt.year).agg({'Passenger Count' : 'sum'})

#print to make sure everything looks correct
print(final_extraction)
```

```python
#graph the extracted data
final_extraction.plot(kind='line', marker='o', figsize=(10, 6))

plt.title('Deplaning for Month of May 2003-2018')
plt.xlabel('Year')
plt.ylabel('Total Passenger Count')

plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```

In order to obtain the necessary extracted data, I first narrowed things down to just the desired years (2003 - 2018) + month (May). After that, I extracted the data for which the "Activity Code" was "Deplane" (as that was the desired value) and while grouping everything by year and then aggregating the passenger counts using the sum. This gave me the desired information that I needed to work with for the problem. Lastly, I plotted this information into a line chart.

Onto analysis!

## Question 6:

(10 pts) Predict the passenger count for May 2019 and May 2020, using simple linear regression. Ensure that you visualize the original data and also include the predictions from the model. Compare the predicted passenger count with the actual passenger count. In your explanation, discuss how close/far your predictions are from the actual values.

```python
#import some stuff to use for linear regression
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#reset the index for final_extraction to flatten the multi-index due to aggregating
final_extraction = final_extraction.reset_index()

final_extraction
```

```python
#take a look at the previously extracted data in a scatter plot
plt.figure()
extracted_plot = sns.regplot(data=final_extraction, x='Activity Period Start Date', y='Passenger Count')
extracted_plot.set_title('Deplaning for Month of May 2003-2018')
extracted_plot.set_xlabel('Year')
extracted_plot.set_ylabel('Total Passenger Count')
extracted_plot.ticklabel_format(style='plain', axis='y')
```

From this linear regression plot, it can be seen that there is a clear distinct line with an upwards trend. Using this line, some predictions can be generated for the years 2019 and 2020 in terms of the number of passengers deplaning in May.

```python
#store explanatory variable (year)
X = final_extraction['Activity Period Start Date'].to_numpy()
X = X.reshape(-1,1)

#store response variable (passenger count)
y = final_extraction['Passenger Count'].to_numpy()
y = y.reshape(-1,1)

#initialize linear regression object
lr_model = LinearRegression(fit_intercept = True)
lr_model = lr_model.fit(X, y)

#print intercept and slope
print('The intercept is: ', lr_model.intercept_[0])
print('The slope is: ', lr_model.coef_[0, 0])
```

Using the equation y = mx + b and the slope/intercept that was found previously, we get the equation:

y = 83953.05x - 166992517.90

^slope and intercept was rounded 2 decimal places

In other words, using this equation, by subsituting the year for x, predictions can be made for y (total number of passengers that deplaned in the year) to create predictions for the year 2019 and 2020.

```python
#make prediction for 2019
year_2019 = np.array([2019]).reshape(-1, 1)
print('2019 prediction: ', lr_model.predict(year_2019))

#make prediction for 2020
year_2020 = np.array([2020]).reshape(-1, 1)
print('2020 prediction: ', lr_model.predict(year_2020))
```

```python
#visualize these predictions
plt.figure()
extracted_plot = sns.regplot(data=final_extraction, x='Activity Period Start Date', y='Passenger Count')
extracted_plot.set_title('Deplaning for Month of May 2003-2018 With Predictions')
extracted_plot.set_xlabel('Year')
extracted_plot.set_ylabel('Total Passenger Count')
extracted_plot.ticklabel_format(style='plain', axis='y')
sns.scatterplot(x = [2019, 2020], y = [2508690.05, 2592643.10], color = 'red')
```

```python
#find actual amount for 2019
passenger_2019actual = airTrafficPassengerStats_df[
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.to_period('Y') == '2019') &
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.month == 5) &
    (airTrafficPassengerStats_df['Activity Type Code'] == 'Deplaned')].agg({'Passenger Count' : 'sum'})
print('Actual number passengers that deplaned in 2019:', passenger_2019actual)

#find actual amount for 2020
passenger_2020actual = airTrafficPassengerStats_df[
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.to_period('Y') == '2020') &
    (airTrafficPassengerStats_df['Activity Period Start Date'].dt.month == 5) &
    (airTrafficPassengerStats_df['Activity Type Code'] == 'Deplaned')].agg({'Passenger Count' : 'sum'})
print('Actual number passengers that deplaned in 2020:', passenger_2020actual)
```

Looking at the linear regression graph for the data with the predictions, it can be seen that the predictions for 2019 and 2020 follow the trend and would theoretically have more and more passengers that deplane in May with each succeeding year. Does reality reflect these predictions? Let's look!
